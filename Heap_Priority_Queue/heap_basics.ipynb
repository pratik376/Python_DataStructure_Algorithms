```python
# Build Min Heap (Heapify)
# Time: O(n), Space: O(1)

A = [-4, 3, 1, 0, 2, 5, 10, 8, 12, 9]

import heapq
heapq.heapify(A)

A
```

```python
# Heap Push (Insert element)
# Time: O(log n)

heapq.heappush(A,4)

A
```

```python
# Heap Pop (Extract min)
# Time: O(log n)

min= heapq.heappop(A)

A, min
```

```python
# Heap Sort
# Time: O(n log n), Space: O(n)
# NOTE: O(1) Space is possible via swapping, but this is complex

def heapsort(arr):

    heapq.heapify(arr)
    n=len(arr)
    new_answer= [0] * n

    for i in range(n):
        elemnt= heapq.heappop(arr)
        new_answer[i]=elemnt
    return new_answer

heapsort([1, -3, 5, 7, 9, 2, 4, 6, 8, 0])

```

```python
# Heap Push Pop: Time: O(log n)
# it will remove first element and insert the latest element

print(A)

heapq.heappushpop(A,99)

print(A)
```

```python
# heapq supports mean heap by default but we can make it to max heap

B = [-4, 3, 1, 0, 2, 5, 10, 8, 12, 9]
n = len(B)

for i in range(n):
    B[i]= - B[i]

heapq.heapify(B)
B
```

```python
larget= -heapq.heappop(B)
larget
```

```python
heapq.heappush(B, -7) # Insert 7 into max heap

B
```

```python
# Build heap from scratch - Time: O(n log n)

C = [-5, 4, 2, 1, 7, 0, 3]

heap=[]

for x in C:
    heapq.heappush(heap,x)
    print(heap, len(heap))
```

```python
# Putting tuples of items on the heap
D = [5, 4, 3, 5, 4, 3, 5, 5, 4]

from collections import Counter
counter= Counter(D)

counter.items()
```

```python
heap= []

for k, v in counter.items():

    heapq.heappush(heap, (k,v))  # here it will put value based on key 

heap
```
